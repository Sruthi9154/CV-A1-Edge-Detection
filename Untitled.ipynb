```python
import cv2
import matplotlib.pyplot as plt

# Open the image
img = cv2.imread('Rblock.jpeg')

# Apply Canny
edges = cv2.Canny(img, 100, 200, 3, L2gradient=True)

plt.figure()
plt.title('canny')
plt.imsave('canny1.png', edges, cmap='gray', format='png')
plt.imshow(edges, cmap='gray')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def roberts_operator(image):
    # Define Roberts Cross kernels
    kernel_x = np.array([[1, 0], [0, -1]])
    kernel_y = np.array([[0, 1], [-1, 0]])

    # Convolve the image with the kernels
    gradient_x = cv2.filter2D(image, -1, kernel_x)
    gradient_y = cv2.filter2D(image, -1, kernel_y)

    # Combine the gradients to get the magnitude
    gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

    return gradient_magnitude

# Read an image
image = cv2.imread('Rblock.jpeg', cv2.IMREAD_GRAYSCALE)

# Apply Roberts operator for edge detection
edges = roberts_operator(image)

# Display the original and the detected edges
plt.subplot(121), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(122), plt.imshow(edges, cmap='gray'), plt.title('Edges (Roberts Operator)')
plt.show()

```
